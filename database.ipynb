```python
import pandas as pd

# Import the causal_claims_beta.csv file into a DataFrame
df = pd.read_csv('causal_claims_beta.csv')

df.head()
```

```python
# Replace spaces in headers with underscores for Neo4j
df.columns = [col.strip().replace(" ", "_") for col in df.columns]

# Convert NaNs to None for Neo4j
df = df.where(pd.notnull(df), None)

print(f"CSV loaded: {df.shape[0]} rows, {df.shape[1]} columns")
df.head()
```

```python
from neo4j import GraphDatabase

URI = 'bolt://192.168.0.206:7687'
AUTH = ('neo4j', 'password')
DB = 'econograph'

try:
    driver = GraphDatabase.driver(URI, auth=AUTH)
    driver.verify_connectivity()
    print("Connected to Neo4j")
except Exception as e:
    print("Error connecting to Neo4j:", e)
```

```python
# Create or merge JELCode nodes
def merge_jel(tx, code, description=None):
    if code is None:
        return
    query = "MERGE (j:JELCode {code: $code})"
    if description:
        query += " SET j.description = $description"
    tx.run(query, code=code, description=description)

# Create Paper nodes
def create_paper(tx, paper_props):
    keys = list(paper_props.keys())
    set_clause = ", ".join([f"p.{k} = ${k}" for k in keys])
    query = f"CREATE (p:Paper) SET {set_clause}"
    tx.run(query, **paper_props)

# Link papers to JEL cause/effect and create CAUSES relationship
def link_paper_jel(tx, paper_id, cause_code, effect_code):
    if cause_code is None or effect_code is None:
        return
    query = """
    MATCH (p:Paper {paper_id: $paper_id})
    MERGE (cause:JELCode {code: $cause})
    MERGE (effect:JELCode {code: $effect})
    MERGE (cause)-[:CAUSES]->(effect)
    MERGE (p)-[:HAS_CAUSE]->(cause)
    MERGE (p)-[:HAS_EFFECT]->(effect)
    """
    tx.run(query, paper_id=paper_id, cause=cause_code, effect=effect_code)
```

```python
with driver.session(database=DB) as session:
    for _, row in df.iterrows():
        # 1. Create Paper node
        paper_props = {col: row[col] for col in df.columns}
        session.write_transaction(create_paper, paper_props)
        
        # 2. Merge JEL nodes
        session.write_transaction(merge_jel, row['jel_cause'])
        session.write_transaction(merge_jel, row['jel_effect'])
        
        # 3. Link Paper to JEL nodes and create CAUSES relationship
        session.write_transaction(link_paper_jel, row['paper_id'], row['jel_cause'], row['jel_effect'])

print("CSV uploaded and graph relationships created in 'econograph' database!")
```

```python
import pandas as pd
import networkx as nx
import plotly.graph_objects as go

# ======================
# 1. Sample your DataFrame
# ======================
# Take a connected subgraph sample of 100 nodes (papers)
sample_size = 100
df_sample = df_causal.head(sample_size)  # or a smarter sampling of connected nodes

# ======================
# 2. Build the NetworkX graph
# ======================
G = nx.DiGraph()

for _, row in df_sample.iterrows():
    paper_node = f"PAPER: {row['Paper_ID']}\n{row['Title']}"
    cause_node = f"CAUSE: {row['Cause_JEL']}\n{row['Cause_Description']}"
    effect_node = f"EFFECT: {row['Effect_JEL']}\n{row['Effect_Description']}"

    # Add nodes with type info
    G.add_node(paper_node, node_type='paper')
    G.add_node(cause_node, node_type='cause')
    G.add_node(effect_node, node_type='effect')

    # Add edges Paper -> Cause -> Effect
    G.add_edge(paper_node, cause_node)
    G.add_edge(cause_node, effect_node)

# ======================
# 3. Generate layout
# ======================
pos = nx.spring_layout(G, seed=42)  # positions for all nodes

# ======================
# 4. Build Plotly traces
# ======================
edge_x = []
edge_y = []

for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x.extend([x0, x1, None])
    edge_y.extend([y0, y1, None])

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=1, color='#888'),
    hoverinfo='none',
    mode='lines'
)

node_x = []
node_y = []
node_text = []
node_color = []

color_map = {'cause': 'orange', 'effect': 'lightblue', 'paper': 'green'}

for node in G.nodes(data=True):
    x, y = pos[node[0]]
    node_x.append(x)
    node_y.append(y)
    node_text.append(node[0])
    node_color.append(color_map.get(node[1]['node_type'], 'gray'))

node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers',
    marker=dict(size=20, color=node_color, line_width=2),
    text=node_text,
    hoverinfo='text'
)

# ======================
# 5. Create figure
# ======================
fig = go.Figure(data=[edge_trace, node_trace])

# Add custom legend manually
for typ, color in color_map.items():
    fig.add_trace(go.Scatter(
        x=[None], y=[None],
        mode='markers',
        marker=dict(size=15, color=color),
        legendgroup=typ,
        showlegend=True,
        name=typ.capitalize()
    ))

fig.update_layout(
    title='Interactive Causal Subgraph',
    showlegend=True,
    hovermode='closest',
    margin=dict(b=20, l=5, r=5, t=40)
)

# ======================
# 6. Save to HTML
# ======================
import plotly.io as pio
pio.write_html(fig, file="causal_subgraph.html", auto_open=True)
```
